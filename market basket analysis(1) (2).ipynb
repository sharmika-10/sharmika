```python
#apriori algorithym
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
data
```

```python
data['Quantity'] = 1
data
```

```python
transactions =  data.groupby(['Member_number','itemDescription'])['Quantity'].sum().unstack().reset_index().set_index('Member_number')
transactions = transactions.fillna(0)
transactions
```

```python
def encode(x):
    if x <=0:
        return 0
    elif x>=0:
        return 1
basket = transactions.applymap(encode)
basket
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Apply Apriori
frequent_itemset = apriori(basket, min_support=0.06, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemset, metric='lift', min_threshold=1)

# Display rules
print(rules.head())
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Apply Apriori
frequent_itemset = apriori(basket, min_support=0.06, use_colnames=True)

# Add num_itemsets for older mlxtend versions
rules = association_rules(
    frequent_itemset, 
    metric='lift', 
    min_threshold=1, 
    num_itemsets=len(frequent_itemset)
)

# Display rules
print(rules.head())
```
